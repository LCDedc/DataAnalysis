```python
from scipy import io
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data1=pd.read_excel("data/Normalmsdata.xlsx",header=None)
data2=pd.read_excel("data/MildNCDmsdata.xlsx",header=None)
data3=pd.read_excel("data/MCImsdata.xlsx",header=None)
data4=pd.read_excel("data/MajorNCDmsdata.xlsx",header=None)
```

```python
data=pd.concat([data1,data2,data3,data4])
data.head()
```

```python
data.columns=["Gfp","OCC","Dur","Cov"]#设置列名
data.head()
```

```python
data.to_excel("data.xlsx",index=False)
```

```python
data=pd.read_excel("data.xlsx")
data.head()
```

```python
list1={1:"Normal",2:"MildNCD",3:"MCI",4:"MajorNCD"}
data.Group=data.Group.map(list1)
```

```python
data.head()
```

```python
# 绘制分组小提琴图
import seaborn as sns
import matplotlib.pyplot as plt
fig,ax=plt.subplots(2,2,constrained_layout=True,figsize=(8,8))
sns.violinplot( y=data["Gfp"], x=data["Group"] ,ax=ax[0,0])
sns.violinplot( y=data["OCC"], x=data["Group"] ,ax=ax[0,1])
sns.violinplot( y=data["Dur"], x=data["Group"] ,ax=ax[1,0])
sns.violinplot( y=data["Cov"], x=data["Group"] ,ax=ax[1,1])
plt.show()
```

```python
data["Mode"]=["A","B","C","D","E"]*77
data.head()
```

```python
fig,ax=plt.subplots(4,4,constrained_layout=True,figsize=(16,16))
sns.violinplot( y=data[data["Group"]=="Normal"]["Gfp"], x=data["Mode"],ax=ax[0,0])
sns.violinplot( y=data[data["Group"]=="MildNCD"]["Gfp"], x=data["Mode"],ax=ax[0,1])
sns.violinplot( y=data[data["Group"]=="MCI"]["Gfp"], x=data["Mode"],ax=ax[0,2])
sns.violinplot( y=data[data["Group"]=="MajorNCD"]["Gfp"], x=data["Mode"],ax=ax[0,3])
sns.violinplot( y=data[data["Group"]=="Normal"]["OCC"], x=data["Mode"],ax=ax[1,0])
sns.violinplot( y=data[data["Group"]=="MildNCD"]["OCC"], x=data["Mode"],ax=ax[1,1])
sns.violinplot( y=data[data["Group"]=="MCI"]["OCC"], x=data["Mode"],ax=ax[1,2])
sns.violinplot( y=data[data["Group"]=="MajorNCD"]["OCC"], x=data["Mode"],ax=ax[1,3])
sns.violinplot( y=data[data["Group"]=="Normal"]["Dur"], x=data["Mode"],ax=ax[2,0])
sns.violinplot( y=data[data["Group"]=="MildNCD"]["Dur"], x=data["Mode"],ax=ax[2,1])
sns.violinplot( y=data[data["Group"]=="MCI"]["Dur"], x=data["Mode"],ax=ax[2,2])
sns.violinplot( y=data[data["Group"]=="MajorNCD"]["Dur"], x=data["Mode"],ax=ax[2,3])
sns.violinplot( y=data[data["Group"]=="Normal"]["Cov"], x=data["Mode"],ax=ax[3,0])
sns.violinplot( y=data[data["Group"]=="MildNCD"]["Cov"], x=data["Mode"],ax=ax[3,1])
sns.violinplot( y=data[data["Group"]=="MCI"]["Cov"], x=data["Mode"],ax=ax[3,2])
sns.violinplot( y=data[data["Group"]=="MajorNCD"]["Cov"], x=data["Mode"],ax=ax[3,3])
ax[0,3].legend(["Normal""MildNCD","MCI","MajorNCD"])
plt.show()
plt.savefig("小提琴图.png",dpi=600)
```
